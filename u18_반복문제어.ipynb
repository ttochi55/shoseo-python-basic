# Unit 18. break, continue 로 반복문 제어하기

### Break

```python
i = 0
while True:
    i += 1
    if i == 5:
        break
print(i)
```

```python
for i in range(1000):
    if i == 100:
        break
print(i)
```

### Continue

```python
i = 0
while i <= 10:
    i += 1
    if (i % 2 == 1):
        continue
    print(i, end=' ')
```

### 3으로 끝나는 숫자만 출력하기

```python
i = 0
while True:
    if i % 10 != 3:
        i += 1
        continue
    if i > 73:
        break
    print(i, end=' ')
```
